```python
import random 
import tkinter as tk
from tkinter import messagebox

def play_game():
    
    # Possible choices for the game
    choices = ["rock","paper","scissors"]
    
    # Get the user's input from the Entry widget and convert to lowercase
    user_input = user_choice.get().lower()
    
    # Check if the user's input is valid
    if user_input not in choices:
        
        # Display an error message if the input is invalid
        messagebox.showerror("Error", "Invalid choice. Please enter 'rock', 'paper', or 'scissors'.")
        return
    
    # Generate a random choice for the computer
    computer_choice_val = random.choice(choices)

    # Update the label to display the computer's choice
    computer_choice.config(text = f"Computer choice: {computer_choice_val}")
    
    # Get and display the result of the game
    result_game = get_result(user_input, computer_choice_val)
    result.config(text = result_game)


def get_result(user_choice, computer_choice):
    
    # Convert choices to lowercase for case-insensitive comparison
    user_choice = user_choice.lower()
    computer_choice = computer_choice.lower()
    
    # Determine the result of the game based on choices
    if user_choice == computer_choice:
        return f"Both chooses same value: {user_choice}. It's a tie!!!"
        
    elif user_choice == "rock":

        if computer_choice == "scissors":
            return "You Win! Rock smashes Scissors"        
        else:
            return "You Lose! Paper covers Rock"
        
    elif user_choice == "paper":
    
        if computer_choice == "rock":  
            return "You Win! Paper covers Rock"   
        else:
            return "You Lose! Scissors cuts Paper"
    
    elif user_choice == "scissors":
    
        if computer_choice == "paper":
            return "You Win! Scissors cuts Paper"
        else:
            return "You Lose! Rock smashes Scissors"
           
    
def reset_game():
    
    # Clear the user input, reset computer's choice label, and clear result label
    user_choice.delete(0, tk.END)   
    computer_choice.config(text="Computer's choice: ")
    result.config(text="")
    
    
def exit_game():
        
    # Destroy the Tkinter window to exit the game
    app.destroy()

    
# Create the main Tkinter window
app = tk.Tk()
app.title("Rock Paper Scissors Game")
app.geometry("400x400")

# Label for user to enter their choice
label = tk.Label(app, text = "Enter your choice: ")
label.pack()

# Entry widget for user to input their choice
user_choice = tk.Entry(app)
user_choice.pack()

# Label to display the computer's choice
computer_choice = tk.Label( app, text = "Computer's choice: ")
computer_choice.pack()

# Button to play the game
play_button = tk.Button(app, text = "Play", command = play_game)
play_button.pack()

# Button to reset the game
reset_button = tk.Button(app, text = "Reset", command = reset_game)
reset_button.pack()

# Button to exit the game
exit_button = tk.Button(app, text = "Exit", command = exit_game)
exit_button.pack()

# Label to display the result of the game
result = tk.Label(app, text = "")
result.pack()

# Start the Tkinter event loop
app.mainloop()
```
